```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
import time
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.action_chains import ActionChains
from bs4 import BeautifulSoup
```

```python
driver = webdriver.Chrome(ChromeDriverManager().install())
```

```python
def last(driver):
    soup = BeautifulSoup(driver.page_source, 'lxml')
    endText = soup.find_all('ol')[-1].find_all('li')[-1].find('h2').text
    #print(endText)
    #.find('h2').text

    val = ["//*[contains(text(), '",endText,"')]"]
    val=''.join(val)
    end = driver.find_element(by=By.XPATH, value =val)

    actions = ActionChains(driver)
    actions.move_to_element(end).perform()
    
    return endText

def up(driver):
    for i in range(5):
        y = driver.execute_script("return window.pageYOffset;")
        script = ["window.scrollTo(0, ", str( y - 200 ),");"]
        driver.execute_script(''.join(script))
        #time.sleep(5)
```

```python
def scroll(driver, timeout):
    scroll_pause_time = timeout

    last_text = ''#last(driver)

    while True:
        last(driver)
        
        # Wait to load page
        time.sleep(scroll_pause_time)
        
        up(driver)
        
        endText = last(driver)
        
        print(endText, "   ,   ", last_text)
        if str(endText) == str(last_text):
            break
        last_text = endText
        
        up(driver)
```

```python
def theSoup(driver):
    scroll(driver, 5)
    soup = BeautifulSoup(driver.page_source, 'lxml')
    
    #driver.close()
    
    return soup
```

```python
def get_articles(soup):
    arts = soup.find_all('a')
    result = [(link.get('href'), link.find('img'), link.find('h2')) for link in arts if link.find('img') is not None or link.find('h2') is not None]
    return result
```

```python
temp = get_articles(soup)
temp
```

```python
aSet = set()
articles = [] 
Images = []
for (a,b,t) in temp:
    if(a not in aSet):
        tempD = {}
        theUrl = "https://www.nytimes.com" + a
        theTitle = "" if t is None else t.text
        tempD['Url'] = theUrl
        tempD['Images'] = [] if b is None else [{"Description":b.get('alt'), "Link":b.get('src'), "ArticleURL":theUrl, "ArticleTitle":theTitle}]
        if tempD['Images']:
            Images.append(tempD['Images'][0])
        articles.append(tempD)
    aSet.add(a)
    
articles
```

```python
import csv

# csv header
fieldnames = ['Description', 'Link', 'ArticleURL', 'ArticleTitle']

with open('ImageFile.csv', 'w') as f:
    writer = csv.DictWriter(f, fieldnames=fieldnames)
    writer.writeheader()
    writer.writerows(Images)
```

```python
import pandas as pd

dataset = pd.read_csv('ImageFile.csv')
dataset.head()
```

```python
import urllib
import numpy as np
import cv2

theImages = []
for imagepath in dataset["Link"]:
    #print(\"Image: \", imagepath)
    req = urllib.request.urlopen(imagepath)
    arr = np.asarray(bytearray(req.read()), dtype=np.uint8)
    im = cv2.imdecode(arr, -1)
    theImages.append(im)
    time.sleep(3)
```

```python
theImages
```

```python
for i in range(len(theImages)):
    if theImages[i] is not None:
        fn = 'images/' + str(i) + '.png'
        #print(fn, theImages[i])
        cv2.imwrite(fn, theImages[i])
```
