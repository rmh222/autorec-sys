```python
import os, os.path
import numpy as np
```

```python
import torch
import clip
from PIL import Image
```

```python
import pandas as pd
dataset = pd.read_csv('week6.csv')
dataset.head(10)
```

```python
dataset.sample(1)
```

```python
import urllib
import numpy as np
import cv2
import time

theImages = []
theKeywords = [] 
sample = dataset.sample(20)
for imagepath, key in zip( sample["ImagePath"], sample['Keywords'] ) :
    #print(\"Image: \", imagepath)
    while key != key: 
        newEntry = dataset.sample(1)
        imagepath = newEntry["ImagePath"]
        key = newEntry['Keywords']
    theImages.append(imagepath)
    theKeywords.append(key)
        
```

```python
import re
re.findall("\w+", theKeywords[0]) 
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
model, preprocess = clip.load("ViT-B/32", device=device)

keys = re.findall("\w+", theKeywords[0]) 
print(keys)
keys.append("person")
image = preprocess(Image.open(theImages[0])).unsqueeze(0).to(device)
text = clip.tokenize(keys).to(device)

with torch.no_grad():
    image_features = model.encode_image(image)
    text_features = model.encode_text(text)
    
    logits_per_image, logits_per_text = model(image, text)
    probs = logits_per_image.softmax(dim=-1).cpu().numpy()

print("Label probs:", probs)  # prints: [[0.9927937  0.00421068 0.00299572]]
```

```python
#theText = ["a person", "a building", "animated", "a vehicle", "a landscape"]
#text = clip.tokenize(theText).to(device)


for f, k in zip( theImages, theKeywords):
    keys = re.findall("\w+", k) 
    #keys.append("person")
    image = preprocess(Image.open(f)).unsqueeze(0).to(device)
    text = clip.tokenize(keys).to(device)

    with torch.no_grad():
        image_features = model.encode_image(image)
        text_features = model.encode_text(text)

        logits_per_image, logits_per_text = model(image, text)
        probs = logits_per_image.softmax(dim=-1).cpu().numpy()

    print("Filename:", f, "    Label probs:", probs)
    print("Label: ", keys[np.where(probs[0] == probs[0].max())[0][0]]) 
    print()
```
