```python
def last(driver):
    soup = BeautifulSoup(driver.page_source, 'lxml')
    endText = soup.find_all('ol')[-1].find_all('li')[-1].find('h2').text
    #print(endText)
    #.find('h2').text

    val = ["//*[contains(text(), '",endText,"')]"]
    val=''.join(val)
    end = driver.find_element(by=By.XPATH, value =val)

    actions = ActionChains(driver)
    actions.move_to_element(end).perform()
    
    return endText

def up(driver):
    for i in range(5):
        y = driver.execute_script("return window.pageYOffset;")
        script = ["window.scrollTo(0, ", str( y - 350 ),");"]
        driver.execute_script(''.join(script))
        #time.sleep(5)
```

```python
def scroll(driver, timeout):
    scroll_pause_time = timeout

    last_text = ''#last(driver)

    while True:
        last(driver)
        
        # Wait to load page
        time.sleep(scroll_pause_time)
        
        up(driver)
        
        endText = last(driver)
        
        print(endText, "   ,   ", last_text)
        if str(endText) == str(last_text):
            break
        last_text = endText
        
        up(driver)
```

```python
def get_articles(soup):
    arts = soup.find_all('a')
    result = []
    for link in arts: 
        im = link.find('img')
        head = link.find('h2') 
        if im is not None and head is not None:
            result.append((link.get('href'), im, head))
        elif im is not None: 
            p =  im.find_parent('article')
            if p is not None:
                head = p.find('h2')
                result.append((link.get('href'), im, head))
    return result
```

```python
def get_articles1(soup):
    arts = soup.find_all('a')
    result = []
    resSet = set()
    for link in arts: 
        head = link.find('h2') 
        if head is not None and head not in resSet:
            result.append((link.get('href'),head))
            resSet.add(head)
        else: 
            p =  link.find_parent('article')
            if p is not None:
                head = p.find('h2')
                if head not in resSet: 
                    result.append((link.get('href'),head))
                    resSet.add(head)
    return result
```

```python
def get_All_Info(subArticles):
    infoPerArticle = []
    threads = min(30, len(subArticles))
    with concurrent.futures.ThreadPoolExecutor(max_workers=threads) as executor:
        for result in executor.map(get_All_Info_Helper, subArticles):
            infoPerArticle.append(result)
    return infoPerArticle

def get_All_Info_Helper(a):
    images = []
    images.append(extract_img_info(a[1]))
    #images.append(a[1])
    newUrl = "https://www.nytimes.com"
    newUrl = newUrl + a[0]
    theArticle = requests.get(newUrl)
    theSoup = BeautifulSoup(theArticle.content, "html.parser")
    title = a[2].text
    text = []
    theBod = theSoup.find_all('section', {'name': re.compile(r'.*articleBody.*')})
    if theBod is None: 
        theBod = theSoup.find_all('div', {'class': re.compile(r'.*container.*')})
    if theBod: 
        theBod = theBod[0]
        allImgs = theBod.findChildren("img")
        for i in allImgs:
            temp = extract_img_info(i)
            images.append(temp)
        text = [t.text for t in theBod.findChildren("p")]
    temp2 = {}
    temp2["ArticleTitle"] = title
    temp2["Images"] = images
    temp2["ArticleURL"] = newUrl
    temp2["ArticleText"] = text
    return temp2



def extract_img_info(im) -> dict:
    text = im.get("alt")
    link = im.get("src")
    if not str(link).startswith("http"):
        link = im.get("data-src")
    temp = {}
    temp["Image_Description"] = text
    temp["Link"] = link
    return temp
```

```python
# The ners we are most interested in
ners_small = ['PERSON', 'EVENT', 'FAC', 'ORG', 'GPE', 'PRODUCT', 'LOC', 'WORK_OF_ART']
```

```python
topRel = []

# initializing a set containing english stopwords
stop_words = set(stopwords.words('english'))


for i in range(0, len(dataset['Article_Text'])):
    sent = dataset.loc[i,'Article_Text']
    
    lemmatized_words = [y.lemma_ for y in NER(sent) if (y.ent_type_ in set(ners_small) or y.pos_ == 'NOUN') and not y.is_stop and y.pos_ != 'PUNCT']
    tem = [ " ".join(lemmatized_words)]
    
    if(len(tem[0].strip())):
        tfidf_vectorizer = TfidfVectorizer(min_df = 0, max_df = 1, ngram_range=(1,3))
        vect = tfidf_vectorizer.fit_transform(tem)
        feature_array = np.array(tfidf_vectorizer.get_feature_names())
        tfidf_sorting = np.argsort(vect.toarray()).flatten()[::-1]

        n = 20
        top_n = feature_array[tfidf_sorting][:n]
        top_n = set(top_n)
        topRel.append(list(top_n))
    else:
        topRel.append("N/A")
```

```python
def cleanArticle(html):
    soup = BeautifulSoup(html, "html.parser") # create a new bs4 object from the html data loaded
    for script in soup(["script", "style"]): # remove all javascript and stylesheet code
        script.extract()
    # get text
    text = soup.get_text()
    # break into lines and remove leading and trailing space on each
    lines = (line.strip() for line in text.splitlines())
    # break multi-headlines into a line each
    chunks = (phrase.strip() for line in lines for phrase in line.split("  "))
    # drop blank lines
    text = '\n'.join(chunk for chunk in chunks if chunk)
    return text

def clean_article(html):
    soup = BeautifulSoup(html)
    for s in soup(['script', 'style']):
        s.decompose()
    return ' '.join(soup.stripped_strings)
```
